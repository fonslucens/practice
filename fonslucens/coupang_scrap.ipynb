```python
browser = webdriver.Chrome('./chromedriver.exe')
```

```python
browser.get('https://www.coupang.com/np/categories/393760')
```

```python
from bs4 import BeautifulSoup
```

```python
soup = BeautifulSoup(browser.page_source,'html.parser')
```

```python
page = soup.select('ul.baby-product-list')
```

```python
page[0]
```

```python
title = page[0].select('div.name')[0].text.strip()
title
```

```python
price = page[0].select('strong.price-value')[0].text.strip()
price
```
